```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('matches.csv')
df2 = pd.read_csv('matches_2.csv')

df = pd.concat([df, df2], axis=0)
```

```python
# drop duplicates based on match_id
df = df.drop_duplicates(subset=['match_id'])
df
```

```python
def onehot_en(data):
    champ1_cols = [f'champ1_{i}' for i in range(1, 138)]
    champ2_cols = [f'champ2_{i}' for i in range(1, 138)]

    all_cols = champ1_cols + champ2_cols

    test_df = pd.DataFrame(0, index=range(len(data)), columns=all_cols)

    for row_idx, row in enumerate(data):
        for hero_pos, hero_idx in enumerate(row):

            if int(hero_idx) > 0 and hero_pos < 5:
                test_df.loc[row_idx, f'champ1_{hero_idx}'] = 1

            elif int(hero_idx) > 0 and hero_pos >= 5:
                test_df.loc[row_idx, f'champ2_{hero_idx}'] = 1
                
    return test_df

```

```python
df = df.drop(columns=['avg_rank_tier'])
```

```python
df = df.dropna()
df
```

```python
df['team'] = pd.concat([df['radiant_team'], df['dire_team']], axis=1).apply(lambda x: ','.join(map(str, x)), axis=1)
df
```

```python
df_encoded = onehot_en(df['team'].str.split(',').apply(lambda x: [int(i) for i in x]))
```
