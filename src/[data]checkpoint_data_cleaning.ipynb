```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('matches.csv')
df2 = pd.read_csv('matches_2.csv')

df = pd.concat([df, df2], axis=0)
```

```python
# drop duplicates based on match_id
df = df.drop_duplicates(subset=['match_id'])
df
```

```python
mmr_above_5000 = df[df['avg_mmr'] > 5000]
mmr_above_5000
```

```python
sorted_df = df.sort_values(by='avg_mmr', ascending=False)
sorted_df
```

```python
num_nan = df.isna().sum().sum()
num_nan
```

```python
df = df.drop(columns=['avg_rank_tier'])
```

```python
df = df.dropna()
df
```

```python
df['team'] = pd.concat([df['radiant_team'], df['dire_team']], axis=1).apply(lambda x: ','.join(map(str, x)), axis=1)
df
```

```python
df = pd.melt(df, id_vars=['match_id', 'radiant_win', 'duration', 'avg_mmr'], value_vars=['radiant_team', 'dire_team'], var_name='team', value_name='heroes')
df
```

```python
# create a boolean mask for rows where 'team' contains 'dire_team'
mask = df['team'].str.contains('dire_team')

# inverse the boolean in 'radiant_win' for rows where the mask is True
df.loc[mask, 'radiant_win'] = np.invert(df.loc[mask, 'radiant_win'].values)
```

```python
df = df.drop(columns=['match_id', 'team'])
```

```python
df = df.rename(columns={'radiant_win': 'win/loss'})
df
```

```python
df[['hero_1', 'hero_2', 'hero_3', 'hero_4', 'hero_5']] = df['heroes'].str.split(',', expand=True)
df = df.drop(columns=['heroes'])
```

```python
df
```

```python
df['win/loss'] = df['win/loss'].replace({True: 1, False: 0})
df
```

```python
df.dropna(inplace=True)
```

```python
def one_hot_encode(x: np.ndarray):
    ls = list(set(x))
    for i in range(len(ls)):
        ls[i] = ls[i].upper()
    ref = sorted(ls)
    dic = {}

    for i in ref:
        ohe = [0] * len(ref)
        ohe[ref.index(i)] = 1
        dic[i] = ohe

    output = []
    for i in x:
        output.append(dic[i])

    return output

hero_1 = df['hero_1'].tolist()
hero_2 = df['hero_2'].tolist()
hero_3 = df['hero_3'].tolist()
hero_4 = df['hero_4'].tolist()
hero_5 = df['hero_5'].tolist()

hero_1_ = one_hot_encode(hero_1)
hero_2_ = one_hot_encode(hero_2)
hero_3_ = one_hot_encode(hero_3)
hero_4_ = one_hot_encode(hero_4)
hero_5_ = one_hot_encode(hero_5)
df['hero_1'] = hero_1_
df['hero_2'] = hero_2_
df['hero_3'] = hero_3_
df['hero_4'] = hero_4_
df['hero_5'] = hero_5_
```

```python
#write csv
df.to_csv('cleaned_data.csv', index=False)
```
