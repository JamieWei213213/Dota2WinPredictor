```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cleaned_data_final.csv')
```

```python
df
```

```python
def is1(input): #function to make wins - one and loses - 0
    if(input == True): #if win
        return 1
    if(input == False): # if lose 
        return 0
```

```python
df['radiant_win'] = df['radiant_win'].apply(is1)
```

```python
df
```

```python
graphed = df.groupby('radiant_win').sum().drop(columns = ['duration','avg_mmr','match_id'])
graphed
```

```python
graphed = graphed.div(graphed.sum(axis = 0),axis = 1) # converted the dataframe to a proportion of wins and loses

graphed = graphed.drop(index = 0)
first_graphed = graphed.iloc[:,:123]
second_graphed = graphed.iloc[:,-123:] # split the dataframe into two seperate dataframes one for each team 

new_names = ['champ_{}'.format(i+1) for i in range(second_graphed.shape[1])] # make the new column names for both df

first_graphed.columns = new_names
second_graphed.columns = new_names # rename the columns to be the same 


df_combined = pd.concat([first_graphed, second_graphed], ignore_index=True) # concat both columns with the same name


```

```python
df_combined = pd.DataFrame(df_combined.sum(axis = 0)).T
df_combined = df_combined / 2 #combine and take the mean of each column 
```

```python

fig, ax = plt.subplots()
df_combined.T.plot.box(ax=ax)
ax.set_ylabel('Winrate')
ax.set_xticks([])
ax.set_xticklabels([])
plt.legend([], [], frameon=False)
# set the x-axis title
ax.set_xlabel('Heroes')
```

```python
team_graph = df.groupby('radiant_win').count()
team_graph
```

```python
x_axis = ['lost','wins']

team_graph = team_graph['match_id']
ax = team_graph.plot.bar()
for i, v in enumerate(team_graph):
    ax.annotate(str(v), xy=(i, v+1), ha='center')

# assign the x-axis labels to the plot and table
ax.set_xticklabels(x_axis)
ax.set_ylabel('Total Games won')
plt.xlabel('Radiant_team')
plt.show()
```
